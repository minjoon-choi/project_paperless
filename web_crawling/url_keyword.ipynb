```python
import os
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
```

```python
#url = 'https://www.naver.com/'
url = 'https://search.naver.com/search.naver'
```

# 검색어 입력

```python
path = r'C:/Users/ethan/Desktop/'
os.chdir(path)
df = pd.read_excel('file.xlsx')
pre_keyword_list = df['keyword'].dropna(inplace=False)
pre_keyword_list.to_list()
```

```python
keyword_list = [x+' 안경' for x in pre_keyword_list ]
keyword_list
```

# search and crawling 

```python
import time
start = time.time() 

region_keyword = []
counter = [] 
views = [] 
view_links = []

for keyword in keyword_list:
    
    # 네이버 검색 시 사용되는 param 값들을 조합하여 request 
    payload = {'where': 'nexearch', 'sm': 'top_hty', 'fbm': '1', 'ie': 'utf8', 'query': keyword }
    r = requests.get(url, params=payload)
    soup = BeautifulSoup(r.text)
    
    pre_counter = []
    pre_region_keyword = [] 
    i = 1
    
    # a 태그의 content를 리스트에 넣음 

    for element in soup.find_all("a", "api_txt_lines total_tit"):
        pre_counter.append(i)
        pre_region_keyword.append(keyword)
        views.append(element.text) 
        i = i +1 

    # list comprehension
    #views = [element.text for element in soup.find_all("a", "api_txt_lines total_tit")]


    # a 태그의 link를 리스트에 넣음 

    for a in soup.find_all("a", "api_txt_lines total_tit", href=True):
        view_links.append(a['href'])

    # list comprehension
    #view_links = [a['href'] for a in soup.find_all("a", "api_txt_lines total_tit", href=True) if a.text]

    counter = counter + pre_counter
    region_keyword = region_keyword + pre_region_keyword
 
```

```python
# DataFrame화 
d = {'region_keyword' : region_keyword, 'counter' : counter, 'views': views, 'view_links': view_links}
df_naver = pd.DataFrame(data=d)

# '다비치'가 title에 포함되었는지 조사 
df_naver['is_davici'] = np.where(df_naver['views'].str.contains("다비치", case=False, na=False), 'Y', 'N')
df_naver
```

```python
print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
```
